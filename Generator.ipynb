# Preprocessing of text file

```python
def generate_freq(text,k):
    

    token={}
    for i in range(len(text)-k):
        X=text[i:i+k]
        Y=text[i+k]
        
        if token.get(X) is None:
            token[X]={}
            token[X][Y]=1
        else:
            if token[X].get(Y) is None:
                token[X][Y]=1
            else:
                token[X][Y]+=1
    return token
    
```

```python
def freq_to_probab(t):
    for i in t.keys():
        s=float(sum(t[i].values()))
        for j in t[i].keys():
            t[i][j]=t[i][j]/s
    return t
    
    
```

# Training our model

```python
def training_markov_chain(text,k):
    t=generate_freq(text,k)
    T=freq_to_probab(t)
    return T
```

```python
model=training_markov_chain(text,k=4)
```

```python
print(model)
```

# Generating Text at Text time

```python
def sample_next(ctx,T,k):
    ctx = ctx[-k:]
    if T.get(ctx) is None:
        return " "
    possible_Chars = list(T[ctx].keys())
    possible_values = list(T[ctx].values())
    
    
    return np.random.choice(possible_Chars,p=possible_values)
```

```python


def predict_new_text(starting_sent,k,maxLen=1000):
    
    sentence = starting_sent
    ctx = starting_sent[-k:]
    
    for ix in range(maxLen):
        next_prediction = sample_next(ctx,model,k)
        sentence += next_prediction
        ctx = sentence[-k:]
    return sentence

```
